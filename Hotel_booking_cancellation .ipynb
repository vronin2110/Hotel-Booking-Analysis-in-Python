```python
calcellation_percentage = df ['is_canceled'].value_counts(normalize= True)
calcellation_percentage
```

```python
#Ploting Graph
plt.figure(figure =(5,4))
plt.title('Reservation Status Count')
plt.bar(['Not Cancelled','Cancelled'],df ['is_canceled'].value_counts(),edgecolor = 'k' ,width =0.7)
plt.show()
```

```python
# from above graph we can see cancellation rate is high
```

```python
plt.figure(figure =(8,4))
ax1 = sns.countplot(x ='hotel',hue='is_canceled',data = df , palette='Blues')
legend_labels,_ = ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title('Reservation table in diffrent Hotels',size = 20)
plt.xlabel('hotel')
plt.ylabel('No of reservation')
plt.legend(['No Cancelled','cancelled'])
plt.show()
```

```python
# So basically our data Set consists of two types of hotel , the above graph shows us cancellation of both the hotels.
#but from the graph we can see city hotels have higher cancellation , so this is the major issue.
```

```python
resort_hotel = df[df['hotel']== 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
city_hotel = df[df['hotel']== 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize = True)
```

```python
resort_hotel =resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel =city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize =(30,8))
plt.title('Avarage Daily Rate in City and Resort',size = 30)
plt.plot(resort_hotel.index,resort_hotel['adr'],label = 'Resort Hotel')
plt.plot(city_hotel.index,city_hotel['adr'],label = 'City Hotel')
plt.legend(fontsize = 20)
plt.show()
```

```python
#Above graph represent average daily rate of city  and  Resort Hotel
```

```python
df['month'] = df['reservation_status_date'].dt.month
plt.figure(figsize =(16,8))
ax1 = sns.countplot(x ='month',hue='is_canceled',data = df , palette='bright')
legend_labels,_ = ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title('Reservation Status Per month',size = 20)
plt.xlabel('month')
plt.ylabel('No of reservation')
plt.legend(['No Cancelled','cancelled'])
plt.show()
```

```python
# The above graph shows us reservation status per month .
# we can see january has highest cancellation
```

```python
adr_per_month =df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index()
```

```python
plt.figure(figsize =(15,8))
plt.title('ADR Per month',size = 30)
plt.bar(adr_per_month['month'],adr_per_month['adr'])
plt.show()
```

```python
# here we can see january has highest adr and also highest calculation
```

```python
canceled_data = df[df["is_canceled"]==1]
#print(canceled_data)
top_10_country =canceled_data['country'].value_counts()[:10]
plt.figure(figsize =(8,8))
plt.title('Top 10 countries')
plt.pie(top_10_country,autopct='%.2f',labels = top_10_country.index)
plt.show()
```

```python
# Above graph shows highest cancellation rate in top 10 countries
```

```python
# here we can see from where our major cancellation occur
canceled_data['market_segment'].value_counts(normalize=True)
```

```python
canceled_data_adr =canceled_data.groupby('reservation_status_date')[['adr']].mean()
canceled_data_adr.reset_index(inplace=True)
canceled_data_adr.sort_values('reservation_status_date' ,inplace=True)
```

```python
not_canceled_data = df[df["is_canceled"]==0]
not_canceled_data_adr =not_canceled_data.groupby('reservation_status_date')[['adr']].mean()
not_canceled_data_adr.reset_index(inplace=True)
not_canceled_data_adr.sort_values('reservation_status_date' ,inplace=True)
```

```python
plt.figure(figsize=(20,6))
plt.title('Average Daily rate')
plt.plot(canceled_data_adr['reservation_status_date'],canceled_data_adr['adr'],label ='Cancelled')
plt.plot(not_canceled_data_adr['reservation_status_date'],not_canceled_data_adr['adr'],label='Not Cancelled')
plt.legend()
```

```python
# above graph shows average daily rate of cancellation and non cancellation bookings 
```
